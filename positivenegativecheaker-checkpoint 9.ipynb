```python
num = int(input("Enter a number between -1000 and 1000: "))

if num < -1000 or num > 1000:
    print("Number is out of range. Please enter between -1000 to 1000.")
else:
    if num == 0:
        print("The number is Zero")

    elif num > 0:
        print(f"{num} is a Positive number")

    else:
        print(f"{num} is a Negative number")

        if num >= -1000 and num <= -1:
            print(f"{num} is in the negative range (-1 to -1000)")
```
